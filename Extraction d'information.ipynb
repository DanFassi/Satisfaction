```python
service_client = ["mail","service client"]
produit = ["produit", "qualité", "prix", "taille"]
livraison = ["livraison","colis","mois","semaines", "jours", "commandes"]
retour_et_remboursement = ["remboursement","retour"]
category_dict = {"service_client" : service_client,"produit" : produit,
                 "livraison" : livraison, "retour_et_remboursement": retour_et_remboursement}
```

```python
def find_category(my_comment, my_category):
    presence = 0
    for word in my_comment.lower().split(" "):
        if word in my_category:
            presence = 1
    return presence            
```

```python
df1 = df
for nom, category in category_dict.items():
    df1[nom] = df1.Commentaires_reduit.apply(lambda row : find_category(row, category))
```

# TF-IDF

Pour extraire les information pertinente des commentaires, on se base sur l'algortihme tf_idf qui va attribuer un score à chaque mot en fonction de son importance par rapport à l'ensemble des commentaires

Pour chaque commentaire nous extrayons les N mots (On a choisit ici arbitrairement 10) ayant le score le plus important. 

Cela nous permet ainsi d'extraire les informations issue des mots clés de chaque commentaires.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

# application de la vectorisation tf_idf et création d'un dictionnaire attribué à chaque mots sont score tfidf.
vectors = vectorizer.fit_transform(df.Commentaires_reduit)
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()

# transformation du dictionnaire en dataframe pour une meilleur exploitation par la suite.
df_vector = pd.DataFrame(denselist, columns=feature_names)
```

```python
display(df_vector)
```

```python
# Création d'une fonction permettant d'extraire à chaque ligne du dataframe de score les n mots associés 
# aux n scores les plus grands

def extraction(i, n):
    extract = list(df_vector.iloc[i,:][(df_vector.iloc[i,:] > 0)].nlargest(n).index)
    return extract
```

```python
#création d'une colonne index pour pouvoir utiliser l'index de chaque ligne dans la fonction lambda
df["index"] = df.index
#application de la fonction definis plus haut à chaque ligne de notre dataframe
df["extract"] = df["index"].apply(lambda row : extraction(row, 10))
```
